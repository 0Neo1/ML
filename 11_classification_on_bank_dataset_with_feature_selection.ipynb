### problem with **pdays** column

```python
# see unique values in "pdays"
bankdata['pdays'].unique()
```

```python
# make box and whisker plot
# source: https://en.wikipedia.org/wiki/Box_plot
# IQR (Interquartile Range): https://en.wikipedia.org/wiki/Interquartile_range
# get all index where is 999
indexs_999=bankdata.index[bankdata['pdays']==999]
# make drop all indexs_999 (the rows having 999)
#bankdata.drop(labels=indexs_999,axis=0)
# or directly
#bankdata.where(bankdata['pdays']!=999).dropna()# drop all values (rows) having value 999
plt.figure(figsize=(10,5))              # make figure
plt.subplot(1,2,1)                      # add subplot 1 (1 row, 2 columns)
sns.boxplot(data=bankdata,y='pdays')    # plot all data
plt.grid()                              # add grid lines
plt.title('all data')                   # plot title
plt.subplot(1,2,2)                      # add subplot 2 indexs_999(1 row, 2 columns)
sns.boxplot(data=bankdata.drop(labels=indexs_999,axis=0),y='pdays')# plot data with removing all 999 from "pday" column
plt.grid()                              # add grid lines
plt.title('data without "999"')         # plot title
plt.show()                              # make show
```

```python
# make copy of data and replace "999" with -1 in "pday" column
bankdata_copy=bankdata.copy()
bankdata_copy['pdays'].replace(to_replace=999,value=-1,inplace=True)
# see again box plot
sns.boxplot(data=bankdata_copy,y='pdays')# plot all data
plt.grid()                              # add grid lines
plt.show()                              # make show
# still not very effective - need normalization !
```

### label encode values - [Label Encoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)

```python
# load label encoder
from sklearn.preprocessing import LabelEncoder

# make dictionary for label encoders
encoders=dict()

for column in bankdata_copy:            # or #for column in bankdata.columns:
  if bankdata_copy[column].dtype==object:# check if column is object type or not
    encoders[column]=LabelEncoder()     # add key (column name) and value (labelencoder object)
    bankdata_copy[column]=encoders[column].fit_transform(bankdata_copy[column])# generate label and transform labels

# decode labels -
# exmaple 3 for job column - encoders['job'].inverse_transform([3])
# i.e. encoders[column].inverse_transform(label_list)

# see encode dataframe
bankdata_copy.head()
```

### [feature selection](https://scikit-learn.org/stable/modules/feature_selection.html)

#### feature importance by correlation ([pandas.DataFrame.corr](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.corr.html))

```python
# calculate co-relation score
corr_score=bankdata_copy.corr()[['y']]
corr_score.T
```

```python
# set threshold - 0.1
corr_score[np.logical_or(corr_score<-0.1,corr_score>0.1)].dropna().T
```

**heatmap**

```python
plt.figure(figsize=(15,10))             # make figure (.2g - 2 significant digits)
sns.heatmap(data=bankdata_copy.corr(),annot=True,fmt='.2f',linewidths=0.5)# plot heatmap
plt.show()                              # make plot show
```

#### feature importance by [chi2](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.chi2.html#sklearn.feature_selection.chi2)

NOT WORK - only for positive values

#### feature importance by [f_classif (f-score)](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.f_classif.html)

**Compute the ANOVA F-value for the provided sampl**e.

```python
# load f-score for classification method
from sklearn.feature_selection import f_classif
# calculate score
score_f,p_value=f_classif(bankdata_copy.iloc[:,:-1],bankdata_copy['y'])
score_f=score_f/score_f.max()# normalize score
score_f# see score
```

#### feature importance by [mutual_info_classif](https://scikit-learn.org/stable/modules/generated/sklearn.feature_selection.mutual_info_classif.html)

```python
# load mutual info for classification method
from sklearn.feature_selection import mutual_info_classif
# calculate score
score_mi=mutual_info_classif(bankdata_copy.iloc[:,:-1],bankdata_copy['y'])
score_mi=score_mi/score_mi.max()# normalize score
score_mi# see score
```

#### convert all to pandas dataframe

```python
# make
scores_df=pd.DataFrame(data={'corelation-score':bankdata_copy.corr()['y'][:-1],
                             'f-score':score_f,
                             'mutual-info-score':score_mi})
# see dataframe
scores_df
```

```python
# make socring equation (an example equation, no benchmark formula)
# absolute(corelation_score) + mutual_info_score + (f_score)/2
total_score=np.abs(scores_df['corelation-score'])+scores_df['f-score']/2+scores_df['mutual-info-score']
# see best seven scoring column names
total_score.sort_values(ascending=False).index[:7]
```

### NOTE - Don't use **duration** column

Outputs highly depends on call duration (as if someone is interested will be spending more time on call). As well as, only after call we know either one is intreasted in investment scheme or not (than classification have no sense).

### split data - train and test (by [sklearn.model_selection.train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html))

```python
# load spliter module
from sklearn.model_selection import train_test_split

# split data (as pandas dataframe and series)
Xtrain,Xtest,ytrain,ytest=train_test_split(bankdata_copy.drop(columns=['y','duration']),
                                           bankdata_copy['y'],test_size=0.33)
# see shape
Xtrain.shape,ytrain.shape,Xtest.shape,ytest.shape
```

### Classification - Linear Model ([by sklearn.linear_model.LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html))

on seven best columns score (by **f-score** (better for linear relations))

```python
# get seven best linear columns (without "duration" column)
linear_columns=scores_df['f-score'].drop(index='duration').sort_values(ascending=False).index[:7]
# see
linear_columns
```

```python
# load model - Logistic Regression
from sklearn.linear_model import LogisticRegression

# make model (object)
lClf=LogisticRegression(max_iter=1000)
# make model train (optimize coefficients and intercep)
lClf.fit(Xtrain[linear_columns],ytrain)

# see - coefficients (2D array, m for each class)
print('Total coefficients counts:',len(lClf.coef_[0]))
print('All coefficients:',lClf.coef_)
print('Intercept:',lClf.intercept_)

# see model score (accuracy)
print('Model score (training):',lClf.score(Xtrain[linear_columns],ytrain))
print('Model score (testing):',lClf.score(Xtest[linear_columns],ytest))
```

### Classification - K-Nearest Neighbors ([by sklearn.linear_model.KNeighborsClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html))

on seven best columns score (over-all)

```python
# get seven overall best columns (without "duration" column)
nonlinear_columns=total_score.drop(index='duration').sort_values(ascending=False).index[:7]
# see
nonlinear_columns
```

```python
# maximum value of k (square root of total rows)
int(np.sqrt(bankdata_copy.shape[0]))
```

```python
# load K Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

# make list to store scores
scores=dict()
# make loop to check for each value of k
for k in range(1,int(np.sqrt(bankdata_copy.shape[0])+1)):
  # print info
  print('\r',end='')
  print('At k =',k,'/',int(np.sqrt(bankdata_copy.shape[0])+1),end='')
  # make model (object)
  kClf=KNeighborsClassifier(n_neighbors=k)
  # train model
  kClf.fit(Xtrain[nonlinear_columns],ytrain)
  # calculate model score
  scores[k]=kClf.score(Xtest[nonlinear_columns],ytest)
print()
# print results
print('Maximum score:',np.max(list(scores.values())),'; is at k =',
      list(scores.keys())[np.argmax(list(scores.values()))])
print('Minimum score:',np.min(list(scores.values())),'; is at k =',
      list(scores.keys())[np.argmin(list(scores.values()))])
print('Mean score:',np.mean(list(scores.values())))
```

### Classification - Decision Trees ([by sklearn.tree.DecisionTreeClassifier](https://scikit-learn.org/stable/modules/generated/sklearn.tree.DecisionTreeClassifier.html))

on seven best columns score (over-all)

```python
# get seven overall best columns (without "duration" column)
nonlinear_columns=total_score.drop(index='duration').sort_values(ascending=False).index[:7]
# see
nonlinear_columns
```

```python
# load Decision Trees
from sklearn.tree import DecisionTreeClassifier
# load grid seacrh CV
from sklearn.model_selection import GridSearchCV
# load warnings
import warnings
# load FitFailedWarning
from sklearn.model_selection._validation import FitFailedWarning

# hyperparmeters dictionary
hyperparmeters={'criterion':['gini','entropy','log_loss'],
                'splitter':['best','random'],'min_samples_split':[2,3,5]}

# make base model (object)
dtreeClf=DecisionTreeClassifier()
# make object for best model
GridCV_dtreeClf=GridSearchCV(dtreeClf,hyperparmeters,cv=None)
# make search for best parameter - # warning FitFailedWarning
with warnings.catch_warnings():# subpress warnings
  warnings.filterwarnings('ignore',category=FitFailedWarning)
  warnings.filterwarnings('ignore',category=UserWarning)
  GridCV_dtreeClf.fit(Xtrain[nonlinear_columns],ytrain)
# make pandas dataframe of cross validation results
cv_results=pd.DataFrame(GridCV_dtreeClf.cv_results_)[['param_splitter','param_criterion',
                                                      'param_min_samples_split',
                                                      'mean_test_score','rank_test_score']]
cv_results.sort_values('rank_test_score',inplace=True)# sort values by model rank
# see head
cv_results.head()
```

```python
# test best model
print('Training score:',GridCV_dtreeClf.best_estimator_.score(Xtrain[nonlinear_columns],
                                                              ytrain))
print('Testing score:',GridCV_dtreeClf.best_estimator_.score(Xtest[nonlinear_columns],
                                                             ytest))
```
