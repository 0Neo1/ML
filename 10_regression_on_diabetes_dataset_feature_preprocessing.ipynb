#### load data

```python
# load module
from sklearn.datasets import load_diabetes

# data (bunch)
diabetes_dataset=load_diabetes()

# see describe dataset
print(diabetes_dataset.DESCR)
```

#### dataset inputs and outputs

```python
# see shape of data (X or inputs or independent)
print('Shape of inputs:',diabetes_dataset.data.shape)# numpy array
# Instances / row / samples: 442, columns / features: 10
# see shape of outputs or targets or depenedent variable
print('Shape of outputs:',diabetes_dataset.target.shape)# numpy array
# see feature names
print('Feature names:',diabetes_dataset.feature_names)# list
```

#### make dataframe from data

```python
# make dataframe from data
diabetes_df=pd.DataFrame(data=diabetes_dataset.data,columns=diabetes_dataset.feature_names)
# see head of data
diabetes_df.head()
```

```python
# add target also as columns
diabetes_df['target']=diabetes_dataset.target
# see tail
diabetes_df.tail()
```

```python
# see shape of dataframe
diabetes_df.shape               # rows and columns
```

```python
# see size of dataframe
diabetes_df.size                # rows * columns
```

```python
# all indexs
diabetes_df.index               # as list or range
```

```python
# all columns
diabetes_df.columns             # as column list (index)
```

```python
# info
diabetes_df.info()              # about columns and row info
```

```python
# datatypes
diabetes_df.dtypes              # as series
```

```python
# stats
diabetes_df.describe()          # as dataframe
```

#### split dataset

```python
# load data spliter
from sklearn.model_selection import train_test_split

# split data (as numpy array) - X (numpy.ndarray - 2D), y (numpy.ndarray - 1D)
# X (inputs / independent variable) -
#         $ diabetes_dataset.data
#    or   $ diabetes_df[diabetes_dataset.feature_names].values
#    or   $ diabetes_df[diabetes_dataset.feature_names].to_numpy()
# y (outputs / dependent variable) -
#         $ diabetes_dataset.target
#    or   $ diabetes_df['target'].values
#Xtrain,Xtest,ytrain,ytest=train_test_split(diabetes_dataset.data,diabetes_dataset.target,
#                                           test_size=0.33)
# split data (from pandas) - dataframe (X), and series (y)
# X (inputs / independent variable) -
#         $ diabetes_df.iloc[:,:-1]
#   or    $ diabetes_df.loc[:,'age':'s6']
#   or    $ diabetes_df.loc[:,diabetes_dataset.feature_names]
#   or    $ diabetes_df[diabetes_dataset.feature_names]
# y (outputs / dependent variable) -
#         $ diabetes_df['target']
#   or    $ diabetes_df.drop(columns='target')
Xtrain,Xtest,ytrain,ytest=train_test_split(diabetes_df.iloc[:,:-1],diabetes_df['target'],
                                           test_size=0.33)
# see type of X (train and test)
print('X dtype:',type(Xtrain),type(Xtest))
print('y dtype:',type(ytrain),type(ytest))

# see shapes
Xtrain.shape,ytrain.shape,Xtest.shape,ytest.shape
```
